# Create Loyalty Flag

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <=40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id' , 'loyalty_flag' , 'order_number']].head(60)
```

```python
# Group the DataFrame by the loyalty_flag and calculate statistics for prices
loyalty_stats = ords_prods_merge.groupby('loyalty_flag')['prices'].describe()

# Display the statistics
print(loyalty_stats)
```

```python
ords_prods_merge.groupby('user_id').agg({'prices':['mean']})
```

# Create Spending_rates flag

```python
ords_prods_merge['max_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.max)
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.loc[ords_prods_merge['max_price'] >= 10, 'Spending_rates'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_price'] < 10, 'Spending_rates'] = 'Low spender'
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.groupby('user_id').agg({'days_since_order':['median']})
```

# Create Order_frequency Flag

```python
ords_prods_merge['median_day_since_order'] = ords_prods_merge.groupby(['user_id'])['days_since_order'].transform(np.max)
```

```python
ords_prods_merge.loc[ords_prods_merge['median_day_since_order'] > 20, 'Order_frequency'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_day_since_order'] > 10) & (ords_prods_merge['median_day_since_order'] <= 20), 'Order_frequency'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_day_since_order'] <= 10, 'Order_frequency'] = 'Frequent customer'
```
